```python
# Dropping columns that contain all null values
df_cleaned = df.drop(columns=df.columns[df.isnull().mean() == 1], axis=1)
```

```python
# For categorical columns, imputating using mode
categorical_cols = df_cleaned.select_dtypes(include='object').columns
for col in categorical_cols:
    df_cleaned[col].fillna(df_cleaned[col].mode()[0], inplace=True)
```

```python
# For numerical columns, using median for imputation
numerical_cols = df_cleaned.select_dtypes(include=['float64', 'int64']).columns
for col in numerical_cols:
    df_cleaned[col].fillna(df_cleaned[col].median(), inplace=True)
```

## Handling Outliers

```python
# Using IQR Method for handling Outliers
Q1 = df_cleaned[numerical_cols].quantile(0.25)
Q3 = df_cleaned[numerical_cols].quantile(0.75)
IQR = Q3 - Q1
```

```python
# Dealing with outlies in numerical Columns
for col in numerical_cols:
    if IQR[col] > 0:  # Ensure column is not constant
        lower_bound = Q1[col] - 1.5 * IQR[col]
        upper_bound = Q3[col] + 1.5 * IQR[col]
        df_cleaned = df_cleaned[(df_cleaned[col] >= lower_bound) & (df_cleaned[col] <= upper_bound)]
```

# Generating Tags form Free text

```python
# Combining Free text columns
df_cleaned['COMBINED_TEXT'] = df_cleaned['CORRECTION_VERBATIM'].fillna('') + ' ' + df_cleaned['CUSTOMER_VERBATIM'].fillna('')
```

```python
import nltk
import string
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
# removing punctuation marks, stopwords
def clean_text(text):
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = text.lower()
    text = ' '.join(word for word in text.split() if word not in stop_words)
    return text
```

```python
df_cleaned['CLEANED_TEXT'] = df_cleaned['COMBINED_TEXT'].apply(clean_text)
```

## Extracting Tags Using TF-IDF method

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Applying TF-IDF
tfidf = TfidfVectorizer(max_features=10) 
tfidf_matrix = tfidf.fit_transform(df_cleaned['CLEANED_TEXT'])
```

```python
tags = tfidf.get_feature_names_out()
print("Generated Tags:", tags)
```

```python
def extract_tags(text, tags):
    return ', '.join([tag for tag in tags if tag in text])
```

```python
# Adding a new column with extracted tags
df_cleaned['TAGS'] = df_cleaned['CLEANED_TEXT'].apply(lambda x: extract_tags(x, tags))
```
